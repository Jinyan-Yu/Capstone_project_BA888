```
df = pd.read_csv('crush_with_weather.csv')
pd.DataFrame.from_records(df)
df.head()
```

```
df.dtypes
```

```
X_int_list = ['Number_of_Vehicles','Total_Nonfatal_Injuries','avg_temp_(f)','avg_dew_point_(f)',
              'avg_humidity_(%)','avg_visibility_(mi)','avg_wind_(mph)', 'high_wind_gust_(mph)']
X_object_list = ['Manner_of_Collision','Most_Harmful_Events','Vehicle_Configuration','Road_Surface_Condition','Ambient_Light','Weather_Condition','At_Roadway_Intersection','Distance_From_Nearest_Exit','Distance_From_Nearest_Landmark']
Y = df[['Crash_Severity']].astype('str')
```

```
## Convert not numeric value to int 
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
X = df[X_object_list]
X = X.apply(lambda col: label.fit_transform(col.astype(str)), axis=0, result_type='expand')


## Copy int value from df
X[X_int_list] = df[X_int_list]


## Copy crash serverity value to Y
Y['Crash_Severity'] = label.fit_transform(Y['Crash_Severity'])
```

```
import statsmodels.api as sm
# X = sm.add_constant(X) # adding a constant
model = sm.OLS(Y, X).fit()
model.summary()
```
